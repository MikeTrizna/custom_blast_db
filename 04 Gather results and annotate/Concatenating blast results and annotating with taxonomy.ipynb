```python
import pandas as pd
import os
```

```python
result_list = []
col_names = ['sequence_id','gb_accession','taxid','percent_id','qcov']
for i in range(1000):
    result_file = os.path.join('blast_results','results_{}.tsv'.format(i+1))
    df = pd.read_csv(result_file, sep='\t',names=col_names)
    result_list.append(df)
```

```python
blast_results = pd.concat(result_list)
blast_results.info()
```

```python
blast_results.head()
```

```python
tax_df = pd.read_csv('new_taxonomy_2020_04_01/rankedlineage.dmp', encoding='utf8', 
                       sep='\t\|\t',
                      names=['taxid','tax_name','species','genus',
                             'family','order','class','phylum',
                             'kingdom','superkingdom'])
tax_df.head()
```

```python
tax_df = tax_df.drop(columns=['superkingdom'])
tax_df.head()
```

```python
tax_df.info()
```

```python
merged = blast_results.merge(tax_df, on='taxid', how='left')
merged.head()
```
